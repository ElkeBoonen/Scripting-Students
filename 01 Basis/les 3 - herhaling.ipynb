## Herhalingsoefening
We gaan een script maken dat cijfers van studenten bijhoudt, een bonus berekent met NumPy en de resultaten presenteert via een pratende koe.
- Stap 1: Installatie - zorg dat je de library cowsay hebt geïnstalleerd (ga even kijken wat die doet op PyPI
- Stap 2: De Data - zorg dat je namen en cijfers kan bewaren
- Stap 3: Cijfers Invoeren
- Stap 4: De NumPy Bonus - Stel dat het examen erg moeilijk was en iedereen 2 punten bonus krijgt.
- Stap 5: Overzicht tonen Print per student: "Student [naam] heeft een [cijfer] behaald."
- Stap 6: De Koe - Bereken het gemiddelde van de eindcijfers en laat de koe via cowsay.cow() vertellen wat het klasgemiddelde is.

```python
# ! py -m pip install cowsay

import numpy as np
import cowsay

namen = ["jan", "piet", "joris", "korneel", "marie"]
cijfers = [5,2,8,6,7]
array = np.array(cijfers)

array += 2
# print(array+2) # dit lukt alleen maar als het een array is!!!
# studenten = {"jan":5, "piet":2, "joris":8, "korneel":6, "marie":7} # in een dictionary is het niet mogelijk om snel +2 te doen!!

for i in range(0,len(namen)):
    print("Student", namen[i], "heeft een", array[i], "behaald")

gemiddelde = np.average(array) # average kunnen we gewichten meegeven
print(gemiddelde)
print(array.mean()) # mean is gewone gemiddelde

cowsay.cow("Het gemiddelde is " + str(gemiddelde))
```

## Slicing [start:end:step]

- inclusieve start
- exclusieve end

```python
array = np.arange(10)
print(array)

print(array[2::])
print(array[2:8:]) # excl 8
print(array[2::2])
print(array[::3])
print(array[::-1])

array[::3] = -10 # om de 3 elementen zet het op -10
print(array)

print(array[::2]) # even elementen
print(array[1::2]) # oneven elementen
```

### Oefening
Je hebt een lijst met 24 temperatuurmetingen ontvangen, één voor elk uur van de dag (van 00:00 tot 23:00). Je doel is om specifieke dagdelen te isoleren en te analyseren met behulp van slicing.

- Maak een lijst genaamd metingen met 24 getallen (bijv. tussen de 5 en 20 graden).
- Gebruik slicing om de volgende sub-lijsten te maken :
  - De Ochtend-rush: Selecteer de temperaturen van de eerste 6 uur van de dag (index 0 tot 6).
  - De Late Avond: Selecteer de laatste 4 metingen van de dag. Eis: Gebruik hiervoor negatieve indexen.
  - De 'Trend' Check: Selecteer elke 3 uur een meting (dus 00:00, 03:00, 06:00, etc.) om een snelle trend te zien. Gebruik hiervoor de stepsize parameter.
  - De Middag-dip: Selecteer de uren tussen 12:00 en 18:00, maar toon deze in omgekeerde volgorde .
- Loop door je 'Trend' selectie en print de waarden.

```python
import random 

temp = [random.randint(5,21) for x in range(25)]  # --> ne lijst
metingen = np.random.randint(5,21,24) # --> array

print(temp)
print(metingen)

ochtend = metingen[:6] # je begint te tellen vanaf 0!!! index 6 --> 7e element!!!
print(ochtend)

avond = temp[-4::]
print(avond)

trend = metingen[::3]
print(trend)
for i in range(0,len(trend)):
    print(i*3, "uur met temperatuur",trend[i])

middag = temp[18:11:-1]
print(middag)
```

### Oefening - De Social Media "Influencer" Auditor
Je bent ingehuurd door een marketingbureau om de prestaties van 50 fictieve influencers te analyseren. Omdat de echte data nog niet beschikbaar is, bouw je een simulator die een volledige audit uitvoert. Je moet de data genereren, berekeningen maken op grote schaal en een gedetailleerd rapport opstellen van de meest waardevolle accounts.

- Data Generatie: Gebruik de Faker library om een lijst van 50 unieke accounts te genereren . Elk account moet een dictionary zijn met een username, real_name en country.
- Statistieken Simuleren: Gebruik NumPy om voor alle 50 accounts in één keer drie aparte arrays te maken :
  - followers: Willekeurige getallen tussen 5.000 en 500.000.
  - likes: Willekeurige getallen tussen 500 en 25.000.
  - comments: Willekeurige getallen tussen 50 en 5.000.
- Analyse: Bereken de Engagement Rate voor elk account met de formule: ((likes + comments) / followers) * 100. Doe dit volledig met NumPy-arrays (geen loops voor de berekening!) .
- Rapportage: Maak een overzicht van de "Top Performers". Dit zijn de accounts met een Engagement Rate van meer dan 3%.
- Sortering: Sorteer deze toplijst van hoogste naar laagste engagement.

```python
# install packages
! py -m pip install Faker
```

```python
from faker import Faker
import numpy as np

fake = Faker()

accounts = []
for i in range(0,50):
    accounts.append((fake.user_name(), fake.name(), fake.country()))

followers = np.random.randint(5000,500000, 50)
likes = np.random.randint(500,25000, 50)
comments = np.random.randint(50,5000, 50)

for index, value in enumerate(accounts):
    print(index, value, followers[index], likes[index], comments[index])
    print(((likes[index] + comments[index]) / followers[index]) * 100)
```
