## Loops

- for ... in : begrensde herhaling
    - in een range(0,10)
    - in 'iets' --> string, collectie, lijst

- while : voorwaardelijke herhaling
    - zorg ervoor dat variabele in voorwaarde aangepast wordt in de lus zelf!!

```python
for i in range(0,10): # exclusive upperbound!!
    print(i, "* 7 = ", i*7)

for karakter in "dinsdag":
    print(karakter)
```

```python
antwoord = input("Geef iets in")
while antwoord.lower() != "x":
    antwoord = input("Geef iets in")

antwoord = input("Geef iets in")
while antwoord != "x".upper() :
    antwoord = input("Geef iets in")

antwoord = input("Geef iets in")
while antwoord not in "xX" :
    antwoord = input("Geef iets in")

antwoord = input("Geef iets in")
while antwoord != "x" or antwoord != "X" :
    antwoord = input("Geef iets in")
```

## Functies

- functie start met def
- functie heeft een naam
- functie heeft parameters tussen de haakjes (je MOET altijd haakjes zetten)
- functie kan iets teruggeven

```python
def TelKlinkers(woord):
    teller = 0
    for karakter in woord:
        if karakter in "aeuioAEUIO":
            teller += 1
    return teller

def TelLetters(woord):
    klinkers = 0
    medeklinkers = 0
    for karakter in woord:
        if karakter in "aeuioAEUIO":
            klinkers += 1
        else:
            medeklinkers +=1
    return medeklinkers, klinkers

def TelKarakters(woord):
    klinkers = 0
    medeklinkers = 0
    cijfers = 0
    for karakter in woord:
        if karakter in "aeuioAEUIO":
            klinkers += 1
        elif karakter in "1234567890":
            cijfers += 1
        else:
            medeklinkers +=1
    return medeklinkers, klinkers, cijfers

print(TelKlinkers("Dit is een test"))
print(TelLetters("Dit is een test"))
print(TelKarakters("Dit is een test 123"))

medeklinkers, klinkers = TelLetters("amai dees is wel zot")
print(medeklinkers)
print(klinkers)
```
