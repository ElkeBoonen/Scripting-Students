## Herhalingsoefening
list = ["the", "cat", "sat", "on", "the", "mat", "the", "cat", "sat"]

- Tel hoe vaak elk woord voorkomt in een dictionary
- Druk elk woord en zijn frequentie af, gesorteerd van meest naar minst frequent
- Hoeveel unieke woorden zijn er?
- Druk het meest voorkomende woord af

Verwachte uitvoer: 
<code>
the: 3
cat: 2
sat: 2
on: 1
mat: 1
unieke woorden: 5
meest voorkomend: the</code>

```python
list = ["the", "cat", "sat", "on", "the", "mat", "the", "cat", "sat"]
dict = {}

for word in list:
    if word in dict.keys():
        dict[word] += 1
    else:
        dict[word] = 1

print(dict)

# https://www.geeksforgeeks.org/python/python-sort-python-dictionaries-by-key-or-value/

asc = {k: v for k, v in sorted(dict.items(), key=lambda item: item[1])}
desc = {k: v for k, v in sorted(dict.items(), key=lambda item: item[1], reverse=True)}

print(asc, desc)
        
```

## Reguliere expressies
- https://regexr.com/ --> zorg altijd dat de regex eerst in orde is vooraleer je naar code gaat
- https://docs.python.org/3/library/re.html

Zorg dat je met 'raw strings' werkt!! --> r of R voor string plaatsen

```python
import re

text = """Om 09:07 staarde Jonas al 23 seconden naar hetzelfde Python-stukje code.
Sarah typte "print('help')" om 09:12, maar haar ogen vielen half dicht.
Ahmed beweerde dat hij "volledig wakker" was, terwijl zijn hoofd 3 keer knikte.
Line keek 17 keer naar de klok in minder dan 5 minuten. Om 10:01 riep iemand zachtjes: "Waarom doet die for-loop dat nu wéér?"
Emma had 2 verschillende sokken aan en merkte het pas tijdens de while-loop.
Pieter schreef per ongeluk "sleeeep()" in plaats van "sleep()".
Iemand zuchtte "regex is toch geen echte taal" om 10:44.
De projector viel 1 keer uit, maar 6 studenten hadden het niet eens door.
Om 11:00 besloot de klas dat koffie officieel deel van de syllabus moest worden."""

regex = r"\d{2}:\d{2}"

match = re.search(regex,text)
print(re.search(regex,text))
print(match.group())

matches = re.findall(regex,text)
print(matches)

matches = re.findall(r"\w+\([a-z\"'0-9A-Z]*\)", text)
print(matches)

parts = re.split(r"[ \,!\?:\d-]", text)
print(parts)

parts = re.split(r"[^a-zA-Z]", text)
print(parts)

new_text = re.sub(r"\d", "DIGIT", text)
print(new_text)
```
