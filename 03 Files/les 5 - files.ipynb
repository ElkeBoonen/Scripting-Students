```python
import os
from pathlib import Path
path = Path.cwd()

os.chdir(Path.home())

for (root,dirs,files) in os.walk(path):
    print("current folder", root)

    for subfolder in dirs:
        print(subfolder)
    
    for file in files:
        print(file)
```

### Oefening
- De extensie is .txt.
- De bestandsnaam bevat minstens één cijfer.
- Print voor elk gevonden bestand de volledige naam en de map waarin het staat (parent).

```python
import os
from pathlib import Path
import re

path = Path.cwd()

os.chdir(Path.home())

for (root,dirs,files) in os.walk(path):
   
    for file in files:
        if re.search(r'\d',file) and file.endswith(".txt"):
            print(file)
```

```python
file = open("test.txt","r+")

print(file.encoding)

line = file.readline()
while line !="":
    print("lees een lijn: ", line)
    file.write("schrijf een lijn \n")
    line = file.readline()
file.close()
```

## Organiseren van files

```python
import os, shutil

# os.mkdir("dit is een folder")
# shutil.move("test.txt", "dit is een folder")
shutil.copy("chinese_news.csv", "dit is een folder")
```

```python
import os, shutil, zipfile

bb = zipfile.ZipFile("Breaking bad.zip")
bb.extractall()
bb.close()

os.remove("Breaking bad.zip")

os.mkdir("BB")
```

```python
for (root,dirs,files) in os.walk(os.getcwd()):
   
    for file in files:
        if file.endswith(".zip"):
            shutil.move(file, "BB")
```

```python

for (root,dirs,files) in os.walk(str(os.getcwd()) + "BB"):
   
    for file in files:
        if file.endswith(".zip"):
            zip = zipfile.ZipFile(file)
            zip.extractall()
            zip.close()

            os.remove(file)
```
